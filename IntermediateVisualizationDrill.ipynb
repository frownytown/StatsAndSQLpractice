```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

purchases = pd.read_csv("purchases.csv")
print(purchases.head())
```

```python
purchases.describe()
```

```python
sns.barplot(purchases.ad_views)\
# This plot doesnt really seem to be very useful as it gives us no insight
```

```python
sns.countplot(purchases.ad_views)
# Using a count plot on a bunch of unique values also seems really useless
```

```python
sns.boxplot(purchases.ad_views)
# THis one is more interesting because it gives us some idea of the amount
# ad views and what one standard deviation of them looks like
```

```python
sns.pointplot(purchases.ad_views)
# Not useful
```

```python
sns.barplot(x=purchases.items_purchased, y=purchases.ad_views)
# This attempts to show a relationship between number of views and
# items purchased and does an ok job. It is not super clear though
```

```python
sns.boxplot(x=purchases.items_purchased, y=purchases.ad_views)
# This attempts the same and does a worse job in my opinion
```

```python
sns.pointplot(x=purchases.items_purchased, y=purchases.ad_views)
# Not useful
```

```python
sns.boxplot(x=purchases.country, y=purchases.ad_views)
# THis is useful for showing ad views by country and we can see some interesting
# conclusions just from this simple plot
```

```python
sns.pointplot(x=purchases.country, y=purchases.ad_views)
# Not useful
```

```python
sns.barplot(x=purchases.country, y=purchases.ad_views)
# Seems to just be worse than the box plot for the same data
```

```python
sns.stripplot(x=purchases.country, y=purchases.ad_views)
# This is interesting because it shows where the data is at relative to
# each other
```

```python
sns.swarmplot(x=purchases.country, y=purchases.ad_views)
# Accomplishes the same kind of goal as the stripplot
```

```python
sns.violinplot(x=purchases.country, y=purchases.ad_views)
sns.stripplot(x=purchases.country, y=purchases.ad_views)
# Kind of an interesting graph but I would say not super useful
```
