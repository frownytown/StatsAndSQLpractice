```python
data = pd.read_csv('UNdata.csv', index_col="Year")
```

```python
plt.scatter(x=data['Quantity'], y=data['Trade (USD)'])
plt.xlabel('Quantity')
plt.ylabel('Trade in USD')
plt.title('Quantity vs Trade')
plt.show()
```

```python
plt.close()
# In this representation I attempted to plot a relationship between quantity
# and trade in dollars. This shows that there seems to be a relationship.
```

```python
plt.scatter(data.index, y=data['Trade (USD)'], color='purple')
plt.xlabel('Year')
plt.ylabel('Trade in USD')
plt.title('Year vs Trade')
plt.show()
```

```python
plt.close()
# In this representation I once again compared Year and trade, but this
# time using a scatter plot. Once again we can see a relationship indicating
# that trade increases as the years go buy, but the outliers are more obviously
# outliers with this representation compared to the last one.
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(data['Quantity'], color='purple')
plt.ylabel('Quantity')
plt.title('Year vs Quantity')

plt.subplot(1, 2, 2)
plt.plot(data['Trade (USD)'], color='green')
plt.ylabel('Trade in USD')
plt.title('Year vs Trade in USD')
plt.show()
```

```python
plt.close()
```

```python
# In this representation I used the subplot feature to compare year vs
# quantity side by side with year vs trade. We can see a fairly linear
# increase in both quantity of goods and trade but with some ugliness
# caused by the outlier data.
```
